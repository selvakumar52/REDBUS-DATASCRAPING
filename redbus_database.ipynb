```python
import pandas as pd
import mysql.connector
from mysql.connector import Error

def connect_mysql():
    """Establish a connection to the MySQL server."""
    try:
        conn = mysql.connector.connect(
            host="localhost",
            user="root",
            password="selva",
            port=3306
        )
        print("Connected to MySQL server")
        return conn
    except Error as e:
        print(f"Error connecting to MySQL: {e}")
        return None

def create_database(conn, database_name):
    """Create a new database if it does not already exist."""
    try:
        cursor = conn.cursor()
        cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database_name}")
        conn.commit()
        print(f"Database '{database_name}' created or already exists.")
    except Error as e:
        print(f"Error creating database: {e}")

def use_database(conn, database_name):
    """Select the database to use."""
    try:
        cursor = conn.cursor()
        cursor.execute(f"USE {database_name}")
    except Error as e:
        print(f"Error selecting database: {e}")

def create_tables(conn):
    """Create tables in the selected database."""
    try:
        cursor = conn.cursor()
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS redbus_details (
                ID INT AUTO_INCREMENT PRIMARY KEY,
                route_name VARCHAR(255) NOT NULL,
                route_link VARCHAR(255) NOT NULL,
                bus_name VARCHAR(255) NOT NULL,
                bus_type VARCHAR(255) NOT NULL,
                departing_time VARCHAR(255) NOT NULL,
                duration VARCHAR(255) NOT NULL,
                reaching_time VARCHAR(255) NOT NULL,
                star_rating FLOAT NULL,
                price FLOAT NULL,
                seats_available VARCHAR(255) NOT NULL
            )
        ''')
        conn.commit()
        print("Table 'redbus_details' created successfully or already exists.")
    except Error as e:
        print(f"Error creating table: {e}")

def insert_data_to_mysql(conn, data):
    """Insert data into the table."""
    insert_query = '''INSERT INTO redbus_details (
                        route_name,
                        route_link,
                        bus_name,
                        bus_type,
                        departing_time,
                        duration,
                        reaching_time,
                        star_rating,
                        price,
                        seats_available)
                      VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)'''
    try:
        cursor = conn.cursor()
        cursor.executemany(insert_query, data)
        conn.commit()
        print("Values inserted successfully.")
    except Error as e:
        print(f"Error inserting data: {e}")

def load_csv_to_dataframe(csv_file_path):
    """Load CSV file into a DataFrame."""
    try:
        df = pd.read_csv(csv_file_path)
        print("CSV loaded successfully")
        print("Columns in DataFrame:", df.columns)  # Print column names
        return df
    except Exception as e:
        print(f"Error loading CSV file: {e}")
        return None

def main():
    """Main function to execute the process."""
    # Define file path and database details
    csv_file_path = 'REDBUS.csv'
    database_name = 'redbus_details'

    # Load CSV data
    df = load_csv_to_dataframe(csv_file_path)
    if df is not None and not df.empty:
        # Strip any leading/trailing spaces from column names
        df.columns = df.columns.str.strip()
        
        # Define expected columns based on CSV file
        expected_columns = ['route_name', 'route_link', 'bus_name', 'bus_type', 'departing_time', 'duration', 'reaching_time', 'star_rating', 'price', 'seats_available']
        
        # Ensure expected columns are present
        missing_columns = [col for col in expected_columns if col not in df.columns]
        if missing_columns:
            print(f"Missing columns: {missing_columns}")
            return
        
        # Prepare data for insertion
        data = df[expected_columns].values.tolist()

        # Connect to MySQL
        conn = connect_mysql()
        if conn:
            create_database(conn, database_name)
            use_database(conn, database_name)
            create_tables(conn)
            insert_data_to_mysql(conn, data)
            conn.close()
        else:
            print("Failed to connect to MySQL.")
    else:
        print("No data to insert.")

if __name__ == "__main__":
    main()
```
